# Cumulative Distribution Function

The CDF is a function which accepts an input, $c$, and gives the probabiliti that a random variable's outcome is less than c:

$$ CDF(c) = P(x \leq c) = \int_{-\infty}^{c} P(X=x) dx $$

Where $P(X=x)$ is the PDF of the random variable

# Intuition Demo

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_cdf(mu, std, c, fill_color='b', verbose=True):
    # build domains (bookkeeping for plots)
    x_lo = min(c, mu - 3 * std)
    x_hi = max(c, mu + 3 * std)
    x = np.linspace(x_lo, x_hi, 100)
    x_cdf = np.linspace(x_lo, c, 100)

    # compute cdf
    cdf = norm.cdf(x=c, loc=mu, scale=std)

    # compute pdf
    p = norm.pdf(x, loc=mu, scale=std)
    p_cdf = norm.pdf(x_cdf, loc=mu, scale=std)
    
    if verbose:
        print(f'for a normal distribution with mean {mu:.2f} std dev: {std:.2f} ...')
        print(f'Prob outcome x is less than {c:.2f} = CDF({c:.2f}) = PDF(x < {c:.2f}) = {cdf:.3f}')

    # plot
    sns.set(font_scale=1.3)
    plt.plot(x, p, color='k', linewidth=3, label='pdf')
    plt.fill_between(x_cdf, p_cdf, 0, alpha=.4, color=fill_color, label=f'CDF(c)={cdf:.3f}')
    plt.legend()
    plt.gcf().set_size_inches(10, 5)
    
    return cdf
```

```python
# mean
mu = 14

# std deviation 
std = 4

# position we compute cdf of
c = 10

plot_cdf(mu=mu, std=std, c=c)
```

## Computing prob x falls within a particular range:

```python
# mean
mu = 14

# std deviation 
std = 4

c_start = 10
c_end = 14

plot_cdf(mu=mu, std=std, fill_color='b', c=c_start)
plot_cdf(mu=mu, std=std, fill_color='g', c=c_end)
```
